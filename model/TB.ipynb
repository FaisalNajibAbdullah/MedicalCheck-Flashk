```python
from sklearn.naive_bayes import GaussianNB
from sklearn import preprocessing
import pandas
import pickle

flname = "dataset.csv"
data_train = pandas.read_csv(flname)
data_train
```

```python
#Create a Gaussian Classifier
model = GaussianNB()

#X = data_train[["A1","A2","A3","A4","A5","A6","A7","A8","A9","A10","A11","A12","A13","A14","A15","A16","A17","A18","A19","A20","A21","A22","A23"]]

X = data_train[data_train.columns[:-1].tolist()]
X
```

```python
Y = data_train["Class"]
Y
```

```python
model.fit(X, Y)
confidence = model.score(X, Y)
confidence
```

```python
pickle.dump(model, open('model.pkl','wb'))
```
